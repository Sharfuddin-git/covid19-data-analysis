```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
df = pd.read_csv('/content/covid_data.csv', parse_dates=['Date'])
print(df.head())
total_cases = df.groupby('Country')['Confirmed'].sum()
print(total_cases)
total_cases.plot(kind='bar',color='Blue')
plt.title("Maximum Confirmed cases per country")
plt.xlabel("Country")
plt.ylabel("Confirmed cases")
plt.show()

plt.figure(figsize=(10, 6)) # Create a new figure for the time series plot
for country in df['Country'].unique():
    cases_per_country = df[df['Country'] == country]
    plt.plot(cases_per_country['Date'],cases_per_country['Confirmed'],label=country)
plt.xlabel("Dates")
plt.ylabel("Confirmed cases")
plt.title("COVID-19 Cases Over Time per Country")
plt.legend()
plt.grid(True)
plt.show()

latest_df = df[df['Date']== df['Date'].max()].copy() # Use .copy() to avoid SettingWithCopyWarning
latest_df['Death_Rate'] = (latest_df['Deaths'] / latest_df['Confirmed']) * 100
latest_df['Recovery_Rate'] = (latest_df['Recovered'] / latest_df['Confirmed']) * 100
x=np.arange(len(latest_df['Country']))
width=0.35
plt.figure(figsize=(10, 6))
plt.bar(x - width/2, latest_df['Death_Rate'], width, label='Death Rate', color='red')
plt.bar(x + width/2, latest_df['Recovery_Rate'], width, label='Recovery Rate', color='green')
plt.xticks(x, latest_df['Country'])
plt.ylabel('Rate (%)')
plt.title('Death vs Recovery Rate (Latest Month)')
plt.legend()
plt.tight_layout()
plt.show()
```
